```python
import pyiron
from pyiron.project import Project
pr = Project('import_and_change_format')
import ase
from ase.io.lammpsdata import write_lammps_data
```

```python
path_to_import = "/store/Terry_project/NiAlN_EDIP_fitting_DFT/DATA"
pr.import_from_path(path=path_to_import, recursive=True)
```

```python
pr.job_table()[0:20]
```

```python
jobs = []
for job in pr.iter_jobs():
    jobs.append(job)
```

```python
def get_structure_and_write_lammps(job, iteration):
    # if iteration < 0:
    #     iteration = len(job['output/generic/steps']) -1
    job_path = '/store/Terry_project/NiAlN_EDIP_fitting_DFT/DATA/' + job.name.replace('_', '-')
    name_base = job.name + '_iter_' + str(iteration) + '_lmp.data'
    name_no_ox = job.name + '_iter_' + str(iteration) + '_no_ox_lmp.data'
    #structure = job.get_structure(iteration_step=iteration)
    structure = job.get_final_structure()
    ase_obj = structure.to_ase()
    write_lammps_data(job_path + '/' + name_base, ase_obj)
    ase_obj_no_ox = ase_obj.copy()
    del ase_obj_no_ox[[atom.index for atom in ase_obj_no_ox if atom.symbol=='N']]
    write_lammps_data(job_path + '/' + name_no_ox, ase_obj_no_ox)

    
```
